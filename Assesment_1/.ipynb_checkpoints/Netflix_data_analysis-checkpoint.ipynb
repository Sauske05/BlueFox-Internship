```python
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Replacing null values with a default value
import numpy as np
df[['director', 'cast', 'country','rating', 'duration']] = df[['director', 'cast', 'country', 'rating', 'duration']].replace(np.nan, "unknown")
```

```python
#Dropping null date rows
df.dropna(subset = ['date_added'],inplace =True) 
```

```python
df.isnull().sum()
```

```python
type_count = df['type'].value_counts()
type_count
```

```python
#Plotting the type of data i.e whether a movie or a tv show
import matplotlib.pyplot as plt
type_count.plot(kind='bar')
plt.show()
```

```python
#Top Country that air netflix
top_countries = df['country'].value_counts().head(10)
top_countries
```

```python
top_countries.plot(kind = 'bar')
plt.show()
```

```python
#Adding year column to the dataframe.
df['year_added'] = df['date_added'].str.split(',').str[1].str.strip().astype('int64')
```

```python
df.info()
```

```python
#Histplot for visualization of release years
import seaborn as sns
sns.histplot(data = df, x = 'release_year', bins = 20, kde = True, color = 'skyblue')
```

```python
#Histplot for visualization of years_added
sns.histplot(data=df, x='year_added', bins=20, kde=True, color='skyblue')
```

```python
directors_count = df['director'].value_counts()
directors_count
```

```python
top_ten_known_directors_count = directors_count.iloc[1:11]
top_ten_known_directors_count
```

```python
top_ten_known_directors_count.plot(kind = 'bar')
plt.show()
```

```python
df['rating'].value_counts()
```

```python
top_ten_rating_types = df['rating'].value_counts().iloc[:10]
```

```python
top_ten_rating_types.plot(kind = 'bar')
plt.show()
```

```python
#Most frequently occuring casts data
cast_data = []
for i in df['cast']:
    names = i.split(',')
    for j in names:
        cast_data.append(j.strip())
len(cast_data)
```

```python
cast_data = pd.Series(cast_data)
most_popular_actors = cast_data.value_counts().iloc[1:11]
most_popular_actors
```

```python
most_popular_actors.plot(kind= 'bar')
plt.show()
```

```python
movie_tvshow_types = []
for i in df['listed_in']:
    names = i.split(',')
    for j in names:
        movie_tvshow_types.append(j.strip())
len(movie_tvshow_types)
```

```python
#Genre of the movies and tv shows
movie_tvshow_types = pd.Series(movie_tvshow_types)
top_show_types = movie_tvshow_types.value_counts().iloc[:10]
top_show_types
```

```python
top_show_types.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.ylabel('')
plt.title('Genre Distribution')
plt.axis('equal') 
plt.show()
```
